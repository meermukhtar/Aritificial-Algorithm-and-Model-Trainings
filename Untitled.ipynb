```python
import matplotlib.pyplot as plt

# Define the animal names and their population ranges
animals = ['Markhor', 'Himalayan Brown Bear', 'Snow Leopard']
populations = [(2500, 3000), (70, 80), (200, 250)]

# Set the colors for each animal
colors = ['gold', 'brown', 'silver']

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bars
for i, (low, high) in enumerate(populations):
    ax.bar(animals[i], high, color=colors[i], alpha=0.8)
    ax.text(i, high + 10, f"{high}", ha='center', va='bottom', fontsize=10)

# Set the x-axis labels and title
ax.set_xlabel('Animal')
ax.set_ylabel('Population')
ax.set_title('Endangered Animal Populations in Pakistan')

# Set the y-axis limits to accommodate the highest population
max_pop = max(max(pop) for pop in populations)
ax.set_ylim(0, max_pop + 100)

# Adjust the spacing between bars
plt.subplots_adjust(bottom=0.2)

# Show the plot
plt.show()
```
