```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Sample data: Replace with your actual data
data = {
    'Date': ['2020-01-01', '2020-03-01', '2020-06-01', '2020-09-01', '2021-01-01', '2021-03-01'],
    'Value': [10, 20, 30, 40, 50, 60]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Ensure the Date column is in datetime format

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value'], marker='o', linestyle='-', color='b', label='Value')

# Formatting the x-axis
plt.gca().xaxis.set_major_locator(mdates.YearLocator(1))  # Major ticks every year
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator(3))  # Minor ticks every 3 months
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format major ticks as Year
plt.gca().xaxis.set_minor_formatter(mdates.DateFormatter('%b %d'))  # Format minor ticks as Month Day

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot')

# Adjust layout for better fitting
plt.tight_layout()

# Show grid for both major and minor ticks
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.5)

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
