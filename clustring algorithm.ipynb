```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

```

```python
# Load the data
file_path = r'C:\Users\xaadk\Downloads\133661471417367643\clusterdata_final.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print(data.head())
```

```python

# Scatter plot of the data
plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c='blue', marker='o', edgecolor='k')
plt.title('Scatter Plot of the Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

```

```python
# Function to plot clusters
def plot_clusters(data, kmeans, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
    plt.title(title)
    plt.xlabel('A')
    plt.ylabel('B')
    plt.show()

```

```python
# K-means with 3 clusters
kmeans_3 = KMeans(n_clusters=3, random_state=42)
kmeans_3.fit(data)
plot_clusters(data, kmeans_3, 'K-Means Clustering with 3 Clusters')
```

```python

# K-means with 4 clusters
kmeans_4 = KMeans(n_clusters=4, random_state=42)
kmeans_4.fit(data)
plot_clusters(data, kmeans_4, 'K-Means Clustering with 4 Clusters')

```

```python
# K-means with 5 clusters
kmeans_5 = KMeans(n_clusters=5, random_state=42)
kmeans_5.fit(data)
plot_clusters(data, kmeans_5, 'K-Means Clustering with 5 Clusters')
```
