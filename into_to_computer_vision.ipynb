```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
def view_random_images(target_dir,target_class):
    #setup target folder and class name
    target_folder=target_dir+target_class
    #takes random images
    rand_image=random.sample(os.listdir(target_folder),1)
    #read images
    img=mpimg.imread(target_folder+"/"+rand_image[0])
    #now plot the image
    plt.imshow(img)
    plt.title(target_class)
    plt.axis("off");
    print(f"image shape:{img.shape}")
    return img
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.random.set_seed(42)
#create baches firs
train_datagen=ImageDataGenerator(rescale=1./255)
valid_datagen=ImageDataGenerator(rescale=1./255)
#giving location of directiry
train_dir="pizza_steak/train"
test_dir="pizza_steak/test"
#giving directory
train_data=train_datagen.flow_from_directory(directory=train_dir,batch_size=32,target_size=(244,244),class_mode="binary",seed=42)
valid_data=train_datagen.flow_from_directory(directory=test_dir,batch_size=32,target_size=(244,244),class_mode="binary",seed=42)
```

```python
model=tf.keras.models.Sequential(
[
    tf.keras.layers.Conv2D(
    10,3,activation="relu",input_shape=(244,244,3)
    ),
    tf.keras.layers.Conv2D(
    10,3,activation="relu"
    ),
    tf.keras.layers.MaxPool2D(pool_size=2,padding="valid"),
    tf.keras.layers.Conv2D(
    10,3,activation="relu"
    ),
    tf.keras.layers.Conv2D(
    10,3,activation="relu"
    ),
     tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1, activation="sigmoid"), 
])
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=["accuracy"]
)
```

```python
history = model.fit(train_data, epochs=5, steps_per_epoch=len(train_data), validation_data=valid_data, validation_steps=len(valid_data))
```
