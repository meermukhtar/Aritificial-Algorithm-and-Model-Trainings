```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
url = 'https://tidesandcurrents.noaa.gov/sltrends/data/8727520_meantrend.csv'
data = pd.read_csv(url,skiprows=4)
# Check for the first few rows to understand the structure of the data 
print(data.head())
```

```python
print(data.columns)
```

```python
X = data[['Year']]
y = data.iloc[:, 2]  # Access the column using its index
```

```python
print(y)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
plt.scatter(X, y)

# Plot the model
plt.plot(X, model.predict(X), color='red')

plt.xlabel('Year')
plt.ylabel('Mean Sea Level')
plt.title('Sea Level Trend at Cedar Key')
plt.show()
```

```python
#Evaluating Model Assumptions

#To evaluate whether our model meets the assumptions of linear regression, we can use the following code:


# Check for linearity
residuals = y - model.predict(X)
plt.scatter(model.predict(X), residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
