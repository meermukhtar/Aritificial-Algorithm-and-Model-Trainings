```python
import pandas as pd

# Load the data
file_path = r'C:\Users\xaadk\Downloads\133661471417367643\regressiondata_final.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Scatter plot of the data
plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c='blue', marker='o', edgecolor='k')
plt.title('Scatter Plot of the Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import numpy as np

# Extract the features and target variable
X = data.iloc[:, 0].values
Y = data.iloc[:, 1].values

# Add a column of ones to X to account for the intercept term
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Calculate the coefficients using the Normal Equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)

# Display the coefficients
print(f"Intercept: {theta_best[0]}, Slope: {theta_best[1]}")
```

```python
# Predictions
Y_pred = X_b.dot(theta_best)

# Scatter plot of the data
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, c='blue', marker='o', edgecolor='k')
plt.plot(X, Y_pred, c='red')
plt.title('Data and Regression Line')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Calculate Mean Squared Error
mse = np.mean((Y - Y_pred) ** 2)
print(f"Mean Squared Error: {mse}")
```
